```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target


# training and testing data
from sklearn.model_selection import train_test_split
#assign test data size 25%
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)


clf = tree.DecisionTreeClassifier(random_state=42, max_depth=3)
clf1 = tree.DecisionTreeClassifier(random_state=42, max_depth=4)
clf2 = tree.DecisionTreeClassifier(random_state=42, max_depth=5)
clf3 = tree.DecisionTreeClassifier(random_state=42, max_depth=6)
clf4 = tree.DecisionTreeClassifier(random_state=42, max_depth=7)
clf5 = tree.DecisionTreeClassifier(random_state=42, max_depth=8)
clf6 = tree.DecisionTreeClassifier(random_state=42, max_depth=9)
clf7 = tree.DecisionTreeClassifier(random_state=42, max_depth=10)

clf = clf.fit(X, y)
clf1 = clf1.fit(X, y)
clf2 = clf2.fit(X, y)
clf3 = clf3.fit(X, y)
clf4 = clf4.fit(X, y)
clf5 = clf5.fit(X, y)
clf6 = clf6.fit(X, y)
clf7 = clf7.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
# testing the model
y_pred = clf.predict(X_test)

#importing accurancy score 
from sklearn.metrics import accuracy_score
#printing the accurancy of the model 
print(accuracy_score(y_pred, y_test))

# importing classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

#repeter le 2 dernier module pour clf,clf1,clf2,clf3...    :)
```
