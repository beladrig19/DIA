```python
# import package
from __future__ import print_function
import pandas as pd
import numpy as np
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
globaldeaths = pd.read_csv('deaths_global.csv')
globalrecovered = pd.read_csv('recovered_global.csv')
globalconfirmed = pd.read_csv('confirmed_global.csv')
casescountry = pd.read_csv('cases_country.csv')
globalconfirmed
```

```python
# data cleaning
globaldeaths.columns = map(str.lower, globaldeaths.columns)
globalrecovered.columns = map(str.lower, globalrecovered.columns)
globalconfirmed.columns = map(str.lower, globalconfirmed.columns)
casescountry.columns = map(str.lower, casescountry.columns)
casescountry.columns
```

```python
globaldeaths = globaldeaths.rename(columns={'province/state' : 'state', 'country/region' : 'country'})
globalrecovered = globalrecovered.rename(columns={'province/state' : 'state', 'country/region' : 'country'})
globalconfirmed = globalconfirmed.rename(columns={'province/state' : 'state', 'country/region' : 'country'})
casescountry = casescountry.rename(columns={'country/region' : 'country'})


```

```python
sorted_casescountry = casescountry.sort_values('confirmed', ascending=False).head(5)
sorted_casescountry
```

```python
def highlight_col(x):
    r = 'background-color: red'
    p = 'background-color: purple'
    g = 'background-color: grey'
    temp_df = pd.DataFrame('', index=x.index, columns=x.columns)
    temp_df.iloc[:,4] = p
    temp_df.iloc[:,5] = r
    temp_df.iloc[:,6] = g
    return temp_df

sorted_casescountry.style.apply(highlight_col, axis=None)
```

```python
import plotly.express as px
```

```python
fig = px.scatter(sorted_casescountry.head(10), x='country_region', y='confirmed', size='confirmed', color='country_region', hover_name='country_region', size_max=60)
fig.update_layout()
fig.show()
```

```python
import plotly.graph_objects as go

def plot_cases_for_country(country):
    labels = ['confirmed', 'deaths']
    colors = ['blue', 'red']
    mode_size = [6,8]
    line_size = [4,5]

    df_list = [globalconfirmed, globaldeaths]

    fig = go.Figure()

    for i, df in enumerate(df_list):
        if country == 'World' or country == 'world':
            x_data = np.array(list(df.iloc[:, 5:].columns))
            y_data = np.sum(np.asarray(df.iloc[:, 5:]), axis=0)
        else:
            x_data = np.array(list(df.iloc[:, 5:].columns))
            y_data = np.sum(np.asarray(df[df['Country/Region']==country].iloc[:, 5:]), axis=0)

        fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines+markers' ,
                                  name=labels[i],
                                  line = dict(color=colors[i], width=line_size[i]),
                                  connectgaps=True,
                                  text = 'Total' + str(labels[i]) + ': '+ str(y_data[-1])

                                  ))    
    
        fig.show()
#plot_cases_for_country('US')
interact(plot_cases_for_country, country='World');    

```

```python
import folium
```

```python
world_map = folium.Map(location=[11,0], tiles='cartodbpositron', zoom_start=2, max_zoom=6, min_zoom=2)

for i in range(len(globalconfirmed)):
    lat = globalconfirmed.iloc[i]['lat']
    long = globalconfirmed.iloc[i]['long']
    if pd.isna(lat) or pd.isna(long):
        continue  # Skip this iteration if lat or long is NaN
    
    folium.Circle(
        location=[lat, long],
        fill=True,
        radius=(int((np.log(globalconfirmed.iloc[i,-1]+1.00001))) + 0.2)*50000,
        fill_color='blue',
        color='red',
        tooltip = "<div style='margin: 0; background-color: black; color: white;'>"+
                    "<h4 style='text-align:center;font-weight: bold'>"+globalconfirmed.iloc[i]['country'] + "</h4>"
                    "<hr style='margin:10px;color: white;'>"+
                    "<ul style='color: white;;list-style-type:circle;align-item:left;padding-left:20px;padding-right:20px'>"+
                        "<li>Confirmed: "+str(globalconfirmed.iloc[i,-1])+"</li>"+
                        "<li>Deaths:   "+str(globaldeaths.iloc[i,-1])+"</li>"+
                        "<li>Death Rate: "+ str(np.round(globaldeaths.iloc[i,-1]/(globalconfirmed.iloc[i,-1]+1.00001)*100,2))+ "</li>"+
                    "</ul></div>",
    ).add_to(world_map)
world_map    
```
