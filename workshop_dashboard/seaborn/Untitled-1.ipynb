```python
titanic =  pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'class', 'embark_town'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.displot(titanic['fare'])
```

```python
sns.jointplot(x='age', y='fare', data=titanic, kind='hex')
```
