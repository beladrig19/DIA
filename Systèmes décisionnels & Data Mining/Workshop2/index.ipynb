```python
# EXAMPLE 1

import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = {'ID':[1,2,3,4,5,6],
       'Onion':[1,0,0,1,1,1],
       'Potato':[1,1,0,1,1,1],
       'Burger':[1,1,0,0,1,1],
       'Milk':[0,1,1,1,0,1],
       'Beer':[0,0,1,0,1,0]}
```

```python
df = pd.DataFrame(data)
df = df[['ID', 'Onion', 'Potato', 'Burger', 'Milk', 'Beer']]
df
```

```python
frequent_itemsets = apriori(df[['Onion', 'Potato', 'Burger', 'Milk', 'Beer' ]], min_support=0.50, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules
```

```python
rules [ (rules['lift'] >1.125)  & (rules['confidence']> 0.8)  ]
```

```python
# EXAMPLE 3

movies = pd.read_csv('movies.csv')
movies.head(10)
```

```python
movies_ohe = movies.drop('genres',axis=1).join(movies.genres.str.get_dummies())
pd.options.display.max_columns=100
movies_ohe.head()
```

```python
# Calculer des statistiques sur les genres de films
# Partie 1 : Statistiques bases sur les colonnes 'genres' d'un DataFrame movies_ohe

# Supprimer les colonnes 'title' et 'movieId', puis compter les occurrences de chaque valeur
stat1 = movies_ohe.drop(['title', 'movieId'], axis=1).apply(pd.value_counts)

# Transposer le DataFrame, supprimer les lignes o la valeur est 0, trier par la valeur 1 (nombre de films), et renommer la colonne
stat1 = stat1.transpose().drop(0, 1).sort_values(by=1, ascending=False).rename(columns={1: 'No. of movies'})

# Partie 2 : Statistiques bases sur les colonnes 'genres' et 'movieId' d'un DataFrame movies

# Ajouter une colonne 'genre_count' contenant le nombre de genres pour chaque film
stat2 = movies.join(movies.genres.str.split('|').reset_index().genres.str.len(), rsuffix='r').rename(columns={'genresr': 'genre_count'})

# Slectionner les films ayant un seul genre, supprimer la colonne 'movieId', regrouper par 'genres' et sommer les valeurs
stat2 = stat2[stat2['genre_count'] == 1].drop('movieId', 1).groupby('genres').sum().sort_values(by='genre_count', ascending=False)

# Fusionner les deux ensembles de statistiques bases sur les genres
stat = stat1.merge(stat2, how='left', left_index=True, right_index=True).fillna(0)

# Convertir la colonne 'genre_count' en entier et renommer la colonne
stat.genre_count = stat.genre_count.astype(int)
stat.rename(columns={'genre_count': 'No. of movies with only 1 genre'}, inplace=True)
```
